```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
import re
import time
import random
```

```python
url = 'https://www.alura.com.br/carreiras/ciencia-de-dados'
```

```python
def get_source(url):
    headers={'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
    req = Request(url=url, headers=headers)
    html = urlopen(req).read()
    return html.decode()


def create_soup(url):
    html = get_source(url)

    # Extrair HTML da página
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
soup = create_soup(url)

nome_carreira = soup.find('h1').text.lower().strip().replace(" ", "_")
cards_cursos = soup.find_all('a', class_ = ['course-card is-link', 'course-card --base-level is-link' ])


lista_cursos = []

for card in cards_cursos:
   temp_dict = {}
   temp_dict['nivel'] = card.find('span', class_= re.compile(r'^course-card__tag [nb].*')).text.strip()
   temp_dict['nome'] = card.find('h5').text
   temp_dict['link'] = card.get('href') 
# Adicionar delay aleatório entre 1 e 3 segundos
   time.sleep(random.uniform(1, 3))
   try:

      temp_dict['carga_horaria'] = create_soup(card.get('href')).find('p', class_ = 'courseInfo-card-wrapper-infos').text.replace('h', '')
   except:
      pass
      
   lista_cursos.append(temp_dict)
```

Gerando a planilha


```python
df = pd.DataFrame(lista_cursos)
df.to_csv(f'../{nome_carreira}.csv', index=False)
```
